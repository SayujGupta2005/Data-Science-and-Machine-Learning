```python
df2.plot.bar()
```

```python
df1.plot.line(y='B',figsize=(10,5),lw=1) #Latest version of pandas auto take sthe index for us so we 
#                     dont need to specify x=df1.index
```

```python
df1.plot.scatter(x='A',y='B')
```

```python
df1.plot.scatter(x='A',y='B',c='C') #Now this is a 3d plot where A and B have their axes and C is shown by colour
```

```python
df1.plot.scatter(x='A',y='B',c='C',cmap='coolwarm')
```
