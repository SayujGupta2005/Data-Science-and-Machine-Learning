```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df
y=cancer['target']
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.9,random_state=101)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(x_train,y_train)
```

```python
predictions=model.predict(x_test)
```

```python
from sklearn.metrics import classification_report,confusion_matrix
```

```python
print(confusion_matrix(y_test,predictions))
print(classification_report(y_test,predictions))
```

```python
# Now in some cases, class 0 has all the values =0 ie precision,recall,f1-score,etc are all 0

# In this case,we need to change the parameters of our SVC(check the ? sign on the blue box,
# which will lead to website where we can see the differetn parameters such as C,gamma,etc) 
# for which we'll try different combinations and take the best ones

# This is done by grid_search 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
param_grid={'C':[0,1,1,10,100,1000],'gamma':[1,0.1,.01,.001,.0001]}
```

```python
grid=GridSearchCV(SVC(),param_grid,verbose=3)
# Verbose(default=0) is just the text output of the description of the process
# Atleast fill some number in it,don't let it be deafult 0, so that we may know if something is happening or not
```

```python
grid.fit(x_train,y_train)
```

```python
grid_predictions=grid.predict(x_test)
```
