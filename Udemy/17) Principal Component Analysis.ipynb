```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer=load_breast_cancer()
# Now the cancer variable is a dictionary
```

```python
type(cancer)
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
print(cancer['target'])
```

```python
df=pd.DataFrame(cancer['data'],columns=cancer['feature_names'])
```

```python
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
```

```python
scaler.fit(df)
```

```python
scaled_data=scaler.transform(df)
```

```python
# PCA
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=2)
```

```python
pca.fit(scaled_data)
```

```python
x_pca=pca.transform(scaled_data)
```

```python
scaled_data.shape
# Our scaled_data had 30 components containing a total of 569 instances
```

```python
x_pca.shape
# Now our data has just 2 components
```

```python
plt.figure(figsize=(8,6))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'],cmap='coolwarm')
plt.xlabel('First Principal Component')
plt.ylabel('Second Principal Component')
```

```python
pca.components_
```

```python
df_comp=pd.DataFrame(pca.components_,columns=cancer['feature_names'])
```

```python
df_comp
```

```python
plt.figure(figsize=(12,6))
sns.heatmap(df_comp,cmap='magma')
```
