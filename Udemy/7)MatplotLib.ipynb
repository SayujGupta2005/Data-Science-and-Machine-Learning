```python
x=np.linspace(0,5,11)
y=x**2
```

# Functional Method to create a matplotlib plot

```python
plt.plot(x,y,)
```

```python
# we can add color to the plot by wrting:
plt.plot(x,y,'r-')
```

```python
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
# create subplot on same canvas 
plt.subplot(1,2,1)
plt.plot(x,y,'r-')
plt.subplot(1,2,2)
plt.plot(y,x)
```

# Object Oriented Method to plot graphs

```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X Label')
axes.set_ylabel('Y Label')
axes.set_title('Set Title')
```

```python
fig=plt.figure()
axes1=fig.add_axes([0.1,0.1,0.8,0.8])#The list means [.1*100=10% of campus from left will be starting pt,
#                                                     .1*100=10% of campus from bootom will be starting pt,
#                                                      the axes will take 80% of canvas size in width,
#                                                       and 80% of canvas size in height]
axes2=fig.add_axes([0.2,0.5,0.4,0.3])
axes1.set_title('Larger Plot')
axes2.set_title('Smaller Plot')
axes1.plot(x,y)
axes2.plot(y,x)

```

```python
fig=plt.figure()
axes1=fig.add_axes([.1,.1,.8,.8])
axes2=fig.add_axes([.5,.1,.4,.3]) #taking axes2 to the bottom-right corner
```

```python
fig,axes=plt.subplots(nrows=2,ncols=3)
# axes.plot(x,y)
plt.tight_layout() #Fixes the overlapping of graphs without it
```

```python
fig,axes=plt.subplots(nrows=1,ncols=2)
# for current_ax in axes:
#     current_ax.plot(x,y) #This is to show that axes can be traversed and we can iterate through them
    
axes[0].plot(x,y)
axes[1].plot(y,x)
```

# Figure Size,Aspect Ratio and DPI

```python
# fig=plt.figure(figsize=(3,2))
fig=plt.figure(figsize=(8,2)) # The width increases
ax=fig.add_axes([0,0,1,1])
ax.plot(x,y)
```

```python
fig,axes=plt.subplots(nrows=2,ncols=1,figsize=(8,2))
axes[0].plot(x,y)
axes[1].plot(y,x)

plt.tight_layout()
```

```python
# Legends

fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label='X squared')
axes.plot(x,x**3,label='X cubed')
axes.legend()

# To prevent overlap of error with graph, we can specify loc

axes.legend(loc=0) #best location is automatically selected
```

# Plot Appearance

```python
fig=plt.figure(figsize=(3,3))
axes=fig.add_axes([0,0,1,1])

# we can add colors and change line thickness
axes.plot(x,y,color='green',marker='o',markersize=10)
axes.plot(x,x,linewidth=20)
# we can also put in rgb hex codes for custom color
axes.plot(y,x,'#FF8C00')
# Transparency can be controlled by alpha
axes.plot(y,y,color='purple',alpha=.5,ls=':')
# linestyle ->  ls= 'steps',':','--','=.'
# Markers -> mark the points on the plot -> marker='o','+',etc 
```
