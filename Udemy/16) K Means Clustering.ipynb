```python
# Cluster unlabeled data in unsupervised learning algorithm
```

```python
# K= No of clusters to form(Each cluster is a group which has similar data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import pandas as pd
%matplotlib inline
```

```python
from sklearn.datasets import make_blobs
```

```python
data=make_blobs(n_samples=200,n_features=2,centers=4,cluster_std=1.8,random_state=101)
```

```python
data
# data[0] is the data
# data[1] are their correct labels(which we know bcoz we made the data ourselves,dividing 200 samples into 4 clusters/labels)
```

```python
data[0]
```

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

```python
from sklearn.cluster import KMeans
# KMeans divides the data into clusters after analysing the data
```

```python
kmeans=KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
# These are the labels that it has given to each data based on its analysis
```

```python
# Now we'll create a scatterplot to compare the kmeans_labels_(the K Means output) and data[1] (the correct labels)
```

```python
fig,(ax1,ax2)=plt.subplots(1,2,sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title('Original')
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')

# Color can be different
```
