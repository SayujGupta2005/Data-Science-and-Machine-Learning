# Plotly and Cufflinks

```python
import numpy as np
import pandas as pd
%matplotlib inline
import cufflinks as cf
from plotly.offline import download_plotlyjs,init_notebook_mode,plot,iplot
import plotly.io as pio
pio.renderers.default = 'colab'
init_notebook_mode(connected=True)
cf.go_offline()
```

```python
# Data
df=pd.DataFrame(np.random.randn(100,4),columns='A B C D'.split())
df.head()
```

```python
df2=pd.DataFrame({'Category':['A','B','C'],'Values':[32,43,50]})
df2
```

```python
df.plot()
```

```python
df2.iplot()
```

```python
df.count().iplot(kind='bar')
```

```python
df.iplot(kind='box')
```

```python
# 3-D Surface Block
```

```python
df3=pd.DataFrame({
    'x':[1,2,3,4,5],
    'y':[10,20,30,40,23],
    'z':[5,4,3,2,1]
})
```

```python
df['A'].iplot(kind='hist')
```

```python
df.iplot(kind='hist')
# We'll get overlapping histograms of all the columns shich we can turn off or on by clicking on the variable
```

```python
df[['A','B']].iplot(kind='spread')
```

```python
df.iplot(kind='bubble',x='A',y='B',size='C')
# The dots cahnge their size based on column C
```
